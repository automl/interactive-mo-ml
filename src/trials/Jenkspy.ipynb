# Binning w/ 15

```python
indicators = ["hv", "sp", "ms", "r2"]
min_clusters = 15
binnerizers = {indicator: None for indicator in indicators}

for indicator in indicators:
    x = np.array([value for key, value in data.items() if indicator in key])
    range_n_clusters = range(min_clusters, len(x))
    max_silhouette = -1
    final_clusters = min_clusters
    for n_clusters in range_n_clusters:
        jnb = JenksNaturalBreaks(n_clusters) # Asking for 4 clusters
        jnb.fit(x) # Create the clusters according to values in 'x'
        new_silhouette = silhouette_score(x.reshape(-1, 1), jnb.labels_)
        if  new_silhouette >= max_silhouette:
            binnerizers[indicator] = jnb
            max_silhouette = new_silhouette
            final_clusters = n_clusters
    print(f"{indicator}: silhouette {max_silhouette} final_clusters {final_clusters}")
```

```python
for indicator in indicators:
    x = np.array([value for key, value in data.items() if indicator in key])

    # Create the figure and axis objects
    fig, ax = plt.subplots()

    # Plot each point with the corresponding color
    # for xi, ci in zip(x, jnb.labels_):
    ax.scatter(x, [0] * len(x), c=binnerizers[indicator].labels_, marker='o')

    # Display the plot
    plt.show()
```
